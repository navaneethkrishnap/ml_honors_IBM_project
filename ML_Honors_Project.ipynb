## Importing the Required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
```

## Importing the Dataset

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv")
```

```python
df.head()
```

```python
df.describe()
```

# Data Preprocessing

```python
df.dtypes
# one hot encoding to convert categorical variables to numerical (such as binary variable)
```

## Transforming categorical variables to binary variables

```python
df_pre = pd.get_dummies(data=df,columns=['RainToday','WindGustDir','WindDir9am','WindDir3pm'])
df_pre
```

```python
# Next replace the RainTomorrow target variable from categorical to binary variable
```

```python
df_pre.replace(['No','Yes'],[0,1],inplace=True)
df_pre[["RainTomorrow"]]
```

# Training and Testing data

```python
df_pre.drop('Date',axis=1,inplace=True)
df_pre = df_pre.astype(float)
```

```python
features = df_pre.drop(columns='RainTomorrow',axis=1)
Y = df_pre['RainTomorrow']
```

# **Linear Regression**

```python
X_train,X_test,y_train,y_test = train_test_split(features,Y,test_size=0.2,random_state=10)
```

```python
LinearReg = LinearRegression()
LinearReg.fit(X_train,y_train)
```

```python
predictions = LinearReg.predict(X_test)
```

```python
linearR_MAE = metrics.mean_absolute_error(predictions,y_test)
linearR_MSE = metrics.mean_squared_error(predictions,y_test)
linearR_R2 = metrics.r2_score(predictions,y_test)
print("Linear Regression MAE :",linearR_MAE)
print("Linear Regression MSE :",linearR_MSE)
print("Linear Regression R2_score :",linearR_R2)
```

```python
tab = {'MAE':[linearR_MAE],'MSE': [linearR_MSE],'R2': [linearR_R2]}
tab = pd.DataFrame(tab)
tab.index = ['Linear Regression']
tab
```

# **K-Nearest Neighbour (KNN)**

```python
KNN = KNeighborsClassifier(n_neighbors=4)
KNN.fit(X_train,y_train)
```

```python
predictions = KNN.predict(X_test)
```

```python
knnAccuracyScore = accuracy_score(y_test,predictions)
knnJaccardIndex = jaccard_score(y_test,predictions,pos_label=0)
knnF1Score = f1_score(y_test,predictions,average='weighted')
print("KNN ACCURACY SCORE :",knnAccuracyScore)
print("KNN JACCARD INDEX :",knnJaccardIndex)
print("KNN F1 SCORE:",knnF1Score)
```

# **DECISION TREE**

```python
tree = DecisionTreeClassifier(criterion='entropy',max_depth=8)
tree.fit(X_train,y_train)
```

```python
predictions = tree.predict(X_test)
```

```python
tree_Accuracy_Score = accuracy_score(y_test, predictions)
tree_JaccardIndex = jaccard_score(y_test, predictions, pos_label=0)
tree_F1_Score = f1_score(y_test, predictions)
print('Tree Accuracy Score :', tree_Accuracy_Score)
print('Tree Jaccard Index :', tree_JaccardIndex)
print('Tree F1 Score :', tree_F1_Score)
```

# **Logisitic Regression**

```python
X_train,X_test,y_train,y_test = train_test_split(features,Y,test_size=0.2,random_state=1)
```

```python
LR = LogisticRegression(solver='liblinear')
LR.fit(X_train,y_train)
```

```python
predictions = LR.predict(X_test)
```

```python
lrAccuScore = accuracy_score(y_test,predictions)
lrJaccIndex = jaccard_score(y_test,predictions)
lrF1Score = f1_score(y_test,predictions)
lrLogLoss = log_loss(y_test,predictions)
print('LR Accuracy Score :',lrAccuScore)
print('LR Jaccard Index :',lrJaccIndex)
print('LR F1 Score :',lrF1Score)
print('LR LogLoss :',lrLogLoss)
```

# **Support Vector Machines SVM**

```python
svm = svm.SVC(kernel='linear')
svm.fit(X_train,y_train)
```

```python
predictions = svm.predict(X_test)
```

```python
svmAccScore = accuracy_score(y_test, predictions)
svmJaccIndex = jaccard_score(y_test, predictions, pos_label=0)
svmF1Score = f1_score(y_test, predictions)
print('SVM Accuracy Score: ', svmAccScore)
print('SVM Jaccard Index: ', svmJaccIndex)
print('SVM F1 Score: ', svmF1Score)
```
